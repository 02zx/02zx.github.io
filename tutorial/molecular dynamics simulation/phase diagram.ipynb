```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def phase_boundary(p,a,b,c,d,e):
    return a+b*p+c*np.log(p)+d/p+e*np.sqrt(p)
    
```

```python
#Ice-liquid phase boundary
#[1]: Phase Diagram of H2O: Thermodynamic Functions of the Phase Transitions of HighPressure Ices



plt.figure(figsize=(4,3),dpi=300)

#I [1]
plt.plot(np.arange(0,2085)/10000,
         phase_boundary(np.arange(0,2085),273.0159,-0.0132,-0.1577,0, 0.1516),'-k')
#III [1]
plt.plot(np.arange(2085,3501)/10000,
         phase_boundary(np.arange(2085,3501), 10.277,0.0265,50.1624,0.5868,-4.3288),'-k')
#V [1]
plt.plot(np.arange(3501,6324)/10000,
         phase_boundary(np.arange(3501,6324), 5.0321,-0.0004,30.9482,1.0018,0),'-k')
#VI [1]
plt.plot(np.arange(6324,22160)/10000,
         phase_boundary(np.arange(6324,22160), 4.2804,-0.0013, 21.8756, 1.0018, 1.0785),'-k')
#VII [1]
plt.plot(np.arange(22160,400000)/10000,
         phase_boundary(np.arange(22160,400000), -1355.42, 0.0018, 167.0609,-0.6633,0),'-k')
#X [1]
plt.plot(np.arange(400000,900000)/10000,
         phase_boundary(np.arange(400000,900000), 0.2524, 0.0019, 0.2795, 0.5, 1.1675),'-k')

#Ice-Ice phase boundary
#I-II [1]
plt.plot(np.arange(1500,2130)/10000,
         phase_boundary(np.arange(1500,2130),  0.2569, 0.1026, 0.2931, 0.5, 0.3789),'-k')
#I-III [1]
plt.plot(np.arange(2085,2130)/10000,
         phase_boundary(np.arange(2085,2130),  18.9177,-0.3447, 124.452, 0.5709, 0),'-k')
#II-III [1]
plt.plot(np.arange(2130,3440)/10000,
         phase_boundary(np.arange(2130,3440),   6.8396, 0.0007, 33.13, 0.5, -0.5162 ),'-k')

#II-V [1]
plt.plot(np.arange(3440,6080)/10000,
         phase_boundary(np.arange(3440,6080),   6.6088,-0.028, 36.2934, 0.0086, 0.7358 ),'-k')

#II-VI [1]
plt.plot(np.arange(6080,6500)/10000,
         phase_boundary(np.arange(6080,6500),   1.0231,-0.3688, 6.2306, 0.501, 30.7367 ),'-k')

#III-V [1]
plt.plot(np.arange(3440,3501)/10000,
         phase_boundary(np.arange(3440,3501),    0.7846, 0.1647,-0.5408, 0.9999,-5.3526 ),'-k')

#V-VI [1]
plt.plot(np.arange(6080,6324)/10000,
         phase_boundary(np.arange(6080,6324),    0.1181, 0.4754, -5.8193, 0.9997, -33.7273 ),'-k')

#VI-VII [1]
plt.plot(np.arange(21200,22160)/10000,
         phase_boundary(np.arange(21200,22160), -47.8507, 0, -389.006, 0.9932, 28.8539 ),'-k')

#VI-VIII [1]
plt.plot(np.arange(20000,21200)/10000,
         phase_boundary(np.arange(20000,21200), -8.2483, 0.0518, -81.5196, 0.9991, 0 ),'-k')

#VII-VIII [1]
plt.plot(np.arange(21200,620000)/10000,
         phase_boundary(np.arange(21200,620000), 696.858,-0.0009,-52.7565, 1.4191, 0.8696 ),'-k')

#VII-X [1]
plt.plot(np.arange(400000,630000)/10000,
         phase_boundary(np.arange(400000,630000),  28.3439, -0.007, 331.927, 1.0001, 0 ),'-k')

#VIII-X [1]
plt.plot(np.arange(620000,635501)/10000,
         phase_boundary(np.arange(620000,635501),  10,-0.0128, 9.9998, 10, 9.9873 ),'-k')

#plt.plot([0.1],[210],'k*')

#plt.xscale('log')
#plt.yscale('log')
plt.xlabel('P (GPa)')
plt.ylabel('T (K)')
#plt.xticks([])
plt.yticks([0,50,100,150,200,250,300,350,400,450])
plt.ylim(0,450)
plt.xlim(0,3)

```
